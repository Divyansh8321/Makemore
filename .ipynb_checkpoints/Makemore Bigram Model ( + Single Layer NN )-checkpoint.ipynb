```python
import torch
```

```python
words = open('names.txt' , 'r').read().splitlines() # Generates a list of names
```

```python
b = {}
for w in words:
    bigram = {}
    chs = ['.'] + list(w) + ['.']
    for ch1 , ch2 in zip(chs, chs[1:]):
        bigram = (ch1,ch2)
        b[bigram] = b.get(bigram,0) + 1  
```

```python
sorted(b.items() , key = lambda kv : -kv[1]  )[:10] # Identifying how many times a bigram appears
```

```python
BM = torch.zeros((27,27) , dtype = torch.int32) # Creating a 27x27 torch tensor which will next represent
                                                # the bigrams generated 
```

```python
chars = sorted(list(set(''.join(words)))) # Finding all the unique alphabets in the names set

stoi = {s:i+1 for i,s in enumerate(chars)} # string to integer conversion
stoi ['.'] = 0

itos = {i:s for s,i in stoi.items()} # integer to string 
```

```python
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1 , ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        BM[ix1,ix2] += 1  
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize = (24,24))
plt.imshow(BM, cmap = 'Blues')

for i in range (27):
    for j in range (27):
        chstr = itos[i] + itos[j]
        plt.text(j,i, chstr , ha = 'center'  , va = 'bottom', color ='gray') 
        plt.text(j,i, BM[i,j].item() , ha = 'center' , va = 'top' , color ='gray') 

# Now the following graph represents the coutns for each of these bigrams
```

```python
g = torch.Generator().manual_seed(2147483647)
t = torch.multinomial((BM[0][:]).float() , num_samples = 1 , replacement = True , generator = g )

# Basically normalizes each value in the row and gets the probability of selection and then selects num_samples 
# times with replacement.
```

```python
P = (BM + 1).float()
# The +1 term is here for smoothening

P /= torch.sum(P, axis = 1, keepdim = True)  # Sums along the row and maintains the dimensions as (27,1)
```

```python
g = torch.Generator().manual_seed(2147483647)

# Broadcasting rules copy this 27 times to make it 27,27 and then allows division 
# (WARNING : keepdim = False doesn't work , since the dimension is squeezed out)

for i in range (10):
    name = []
    ix = 0 # starting the new name with the first letter from the first row.
    
    while True:
        p = P[ix]
        ix = torch.multinomial(p, num_samples = 1, replacement = True , generator = g).item()
        name.append(itos[ix])
        if ix == 0:
            break
    print (''.join(name))      
```

```python
# We need a loss function to get an idea of how well the model is performing:

# We will find the probability that the word predicted next is correct

log_likelihood = 0 
n = 0 
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1 , ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1,ix2]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n += 1
avg_log_likelihood = -log_likelihood/n
print (avg_log_likelihood.item())
    
```

```python
# What we want is that the model to correctly predict the next word so we want these probabilities to be as 
# close to 1 as possible. So we will take the product of all these probabilites ('Likelihood') and since that
# will be small we will take log. Now taking log of a product is as good as summing up the indiivual logs ; On 
# adding we will see that the sum is -ve and since we will try to minimise the log function , so we will take -ve.
```

```python
# Now we will try and do this via a neural network
```

```python
xs = [] 
ys = []
num = 0
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1 , ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)
        num += 1
        
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
g = torch.Generator().manual_seed(2147483647)
W = torch.randn(27,27 , requires_grad = True , generator = g) # These represent the log of the counts (logits) of the number of times that 
# particular bigram has appeared ; this is the reason why we chose 27 neurons here.
```

```python
# gradient descent

for k in range(200):

    # forward_pass

    xin = torch.nn.functional.one_hot(xs , num_classes = 27).float() # one-hot encoding 

    logits = xin @ W  # for each input xin it will select those x's as the first letter and then give us the logits of 
    # the bigrams starting with xs and these will be basically normalized and then using these loss will be calculated

    counts = logits.exp()
    probs = counts/ torch.sum(counts ,axis = 1 , keepdim = True)

    # If you notice carefully both these functions are exactly implementing the softmax actiavtion function 

    # the loss for this word emma can be calulated as :
    loss = -probs[torch.arange(num), ys].log().mean() + 0.1 * (W**2).mean() 
    
    # The W**2 term is the regularization term and the constant there represents the extent of smoothening 
    # To think of this , think if all the W were 0 , everything will be exactly uniform 
    # So the more we incentivize W to be 0 , the more uniform the distribution becomes and this is smoothening!
    
    
    
    # backward_pass

    W.grad = None # setting the gradient back to zero before each pass
    loss.backward()
    
    # update :
    W.data += -10 * W.grad
    
print (loss.item()) 
```

```python
# You will notice that it stops at around 2.46 and this means that the weights representing the logits are 
# intiially randomized but eventually we are training them to become same as the initial bigram model we had 
# built (would occur at the same loss)
 
g = torch.Generator().manual_seed(2147483647)

for i in range (10):
    name = []
    ix = 0 # starting the new name with the first letter from the first row.
    
    while True:
        
        xin = torch.nn.functional.one_hot(xs , num_classes = 27).float() # one-hot encoding 
        logits = xin @ W
        counts = logits.exp()
        p = counts/ torch.sum(counts ,axis = 1 , keepdim = True)
        
        ix = torch.multinomial(p[ix], num_samples = 1, replacement = True , generator = g).item()
        name.append(itos[ix])
        if ix == 0:
            break
    print (''.join(name))      

```
